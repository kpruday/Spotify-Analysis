```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('tracks.csv')
data.head()
```

```python
# null values: number of missing values
pd.isnull(data).sum()
```

```python
data.info()
```

```python
sorted_data = data.sort_values('popularity', ascending = True).head(10)
sorted_data
```

```python
data.describe().transpose()
```

```python
most_popular = data.query('popularity>90', inplace = False).sort_values('popularity', ascending = False)
most_popular[:10]
```

```python
data.set_index("release_date", inplace = True)
data.head()
```

```python
data[['artists']].iloc[18]
```

```python
data['duration']=data['duration_ms'].apply(lambda x: round(x/1000))
data.drop('duration_ms', inplace=True, axis=1)
```

```python
data.duration.head()
```

```python
# Check for non-numeric values in each column
non_numeric_values = data.applymap(lambda x: not isinstance(x, (int, float)))
print(non_numeric_values.any())
```

```python
corr_df=data.drop(["key","mode","explicit","id","name","artists","id_artists"],axis=1).corr(method="pearson")
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_df,annot=True,fmt=".1g",vmin=-1,vmax=1,center=0,cmap="inferno",linewidth=1,linecolor="Black")
heatmap.set_title("Correlation HeatMap Between Variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

```python
sample_df=data.sample(int(0.004*len(data)))
```

```python
print(len(sample_df))
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, y="loudness", x="energy", color="c").set(title = "Loudness Vs Energy Correlation")
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, y="popularity", x="acousticness", color="b").set(title = "Popularity Vs Accousticness Correlation")
```

```python
data['dates'] = data.index.get_level_values('release_date')
data['dates'] = pd.to_datetime(data['dates'], errors='coerce', format='%Y-%m-%d')
years_mask = data['dates'].isnull()  
data.loc[years_mask, 'dates'] = pd.to_datetime(data.loc[years_mask, 'dates'], errors='coerce', format='%Y')
years = data['dates'].dt.year
```

```python
sns.displot(years,discrete=True,aspect=2,height=5,kind="hist").set(title="Number of songs per year")
```

```python
data1 = pd.read_csv('SpotifyFeatures.csv')
data1.head()
```

```python
plt.title("Duration of the songs in Different Genres")
sns.color_palette("Set2", as_cmap = True)
sns.barplot(y='genre', x='duration_ms', data=data1)
plt.xlabel("Duration in milli seconds")
plt.ylabel("Genres")
```

```python
sns.set_style(style="darkgrid")
plt.figure(figsize=(10,5))
famous=data1.sort_values("popularity", ascending =False).head(10)
sns.barplot(y='genre',x='popularity',data=famous).set(title="Top 5 Genres by Popularity")
```
